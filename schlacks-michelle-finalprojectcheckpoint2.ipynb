```python
import pandas as pd
import matplotlib 
import matplotlib.pyplot as plt

import numpy as np
%matplotlib inline
import ipywidgets
from ipywidgets import interactive
from datetime import datetime
import bqplot 




```

```python
@ipywidgets.interact(style=plt.style.available)
def make_plot(style):
    with plt.style.context(style): # this style is applied ONLY to this plot and not following
        fig, ax = plt.subplots(1,1,figsize=(6,3))
        ax.scatter(df["Age"], df["Fatalities"], marker= '+')
        ax.set_xlabel("Age")
        ax.set_ylabel("Fatalities")
        ax.set_yscale('linear')
        ax.set_xscale('linear')



    plt.show()
    
    
#interactive_plot = interactive(make_plot)
#output = interactive_plot.children[-1]
#output.layout.height = '350px'
#interactive_plot
```
