What makes me sad is that the youngest person to have died in a shooting was 13 years old. Someone that has not even made it to highschool yet was shot and killed. 

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import ipywidgets
 
df = pd.read_csv('shootings.csv', encoding = "unicode_escape")

@ipywidgets.interact(style=plt.style.available)
def make_plot(style):
    with plt.style.context(style): # this style is applied ONLY to this plot and not following
        fig, ax = plt.subplots(1,1,figsize=(15,5))
        ax.scatter(df["Age"].sort_values(), df["Fatalities"].sort_values(), marker= '+')
        ax.set_xlabel("Age")
        ax.set_ylabel("Fatalities")
        
       


    plt.show()
    
    
```
